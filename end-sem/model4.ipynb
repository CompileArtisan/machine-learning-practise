```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, explained_variance_score
from sklearn.linear_model import LinearRegression
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import LSTM, Dense, Dropout, Input, MultiHeadAttention, LayerNormalization
from tensorflow.keras.callbacks import EarlyStopping
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
import warnings
warnings.filterwarnings('ignore')

def load_and_prepare_data(filepath, target_col='ptp_vel', test_size=0.2, scale_data=True):
    """
    Load and prepare data for time series forecasting
    
    Args:
        filepath: Path to the CSV file
        target_col: Target column name for forecasting
        test_size: Proportion of data to use for testing
        scale_data: Whether to scale features
    
    Returns:
        Dictionary containing processed data components
    """
    print(f"Loading data from {filepath}")
    
    try:
        # Load data
        df = pd.read_csv(filepath)
        print(f"Loaded {len(df)} records with {len(df.columns)} features")
        
        # Check if timestamp column exists
        has_timestamp = 'timestamp' in df.columns
        if has_timestamp:
            # Convert timestamp to datetime if it exists
            df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
            df = df.dropna(subset=['timestamp'])
            print(f"Time range: {df['timestamp'].min()} to {df['timestamp'].max()}")
            
            # Sort by timestamp
            df = df.sort_values('timestamp')
        
        # Basic data checks
        print(f"Target column '{target_col}' stats: Min: {df[target_col].min():.6f}, Max: {df[target_col].max():.6f}")
        missing_count = df[target_col].isna().sum()
        if missing_count > 0:
            print(f"Warning: {missing_count} missing values in target column")
            df = df.dropna(subset=[target_col])
        
        # Select features (exclude non-numeric and identifier columns)
        exclude_cols = ['event_name', 'station_id', 'channel', 'timestamp']
        feature_cols = [col for col in df.columns 
                      if col not in exclude_cols + [target_col] 
                      and df[col].dtype != 'object']
        
        print(f"Selected {len(feature_cols)} features for modeling")
        
        # Split data
        if has_timestamp:
            # Time-based split
            split_idx = int(len(df) * (1 - test_size))
            train_df = df.iloc[:split_idx]
            test_df = df.iloc[split_idx:]
            
            X_train = train_df[feature_cols]
            y_train = train_df[target_col]
            X_test = test_df[feature_cols]
            y_test = test_df[target_col]
            
            if has_timestamp:
                train_dates = train_df['timestamp']
                test_dates = test_df['timestamp']
        else:
            # Random split
            X = df[feature_cols]
            y = df[target_col]
            X_train, X_test, y_train, y_test = train_test_split(
                X, y, test_size=test_size, random_state=42, shuffle=True
            )
            train_dates = None
            test_dates = None
        
        # Scale features if requested
        if scale_data:
            scaler = StandardScaler()
            X_train_scaled = scaler.fit_transform(X_train)
            X_test_scaled = scaler.transform(X_test)
            
            # Create target scaler for easier inverse transform later
            target_scaler = StandardScaler()
            y_train_scaled = target_scaler.fit_transform(y_train.values.reshape(-1, 1)).flatten()
            y_test_scaled = target_scaler.transform(y_test.values.reshape(-1, 1)).flatten()
        else:
            X_train_scaled = X_train
            X_test_scaled = X_test
            y_train_scaled = y_train
            y_test_scaled = y_test
            scaler = None
            target_scaler = None
        
        return {
            'df': df,
            'X_train': X_train, 'y_train': y_train,
            'X_test': X_test, 'y_test': y_test,
            'X_train_scaled': X_train_scaled, 'y_train_scaled': y_train_scaled,
            'X_test_scaled': X_test_scaled, 'y_test_scaled': y_test_scaled,
            'feature_cols': feature_cols,
            'target_col': target_col,
            'train_dates': train_dates,
            'test_dates': test_dates,
            'scaler': scaler,
            'target_scaler': target_scaler,
            'has_timestamp': has_timestamp
        }
    
    except Exception as e:
        print(f"Error loading or preparing data: {e}")
        return None

def exponential_smoothing_forecast(data_dict, forecast_periods=24, seasonal_periods=12):
    """
    Perform Exponential Smoothing forecasting
    
    Args:
        data_dict: Dictionary with prepared data
        forecast_periods: Number of periods to forecast into the future
        seasonal_periods: Number of periods in a seasonal cycle
    
    Returns:
        Dictionary with model, predictions, and metrics
    """
    print("\nPerforming Exponential Smoothing Forecasting:")
    
    y_train = data_dict['y_train']
    y_test = data_dict['y_test']
    
    if len(y_train) < seasonal_periods:
        print(f"Warning: Training data length ({len(y_train)}) is less than seasonal_periods ({seasonal_periods})")
        seasonal_periods = min(seasonal_periods, len(y_train) // 2)
        print(f"Reducing seasonal_periods to {seasonal_periods}")
        
    try:
        # If we have timestamp data, convert to time series
        if data_dict['has_timestamp']:
            train_series = pd.Series(y_train.values, index=data_dict['train_dates'])
            test_series = pd.Series(y_test.values, index=data_dict['test_dates'])
        else:
            train_series = pd.Series(y_train.values)
            test_series = pd.Series(y_test.values)
        
        # Fit Exponential Smoothing model
        model = ExponentialSmoothing(
            train_series,
            seasonal='add',
            seasonal_periods=seasonal_periods
        ).fit()
        
        # Predict test set
        if data_dict['has_timestamp']:
            y_pred = model.forecast(steps=len(test_series))
        else:
            y_pred = model.forecast(steps=len(test_series))
        
        # Calculate metrics
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        mae = mean_absolute_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        
        print(f"ES Model Metrics - RMSE: {rmse:.4f}, MAE: {mae:.4f}, RÂ²: {r2:.4f}")
        
        # Generate future forecast
        future_forecast = model.forecast(steps=forecast_periods)
        
        # Visualize results
        plt.figure(figsize=(12, 6))
        plt.plot(train_series.values, label='Training Data')
        plt.plot(np.arange(len(train_series), len(train_series) + len(test_series)), test_series.values, label='Actual Test Data')
        plt.plot(np.arange(len(train_series), len(train_series) + len(test_series)), y_pred, label='ES Forecast', color='red')
        plt.title('Exponential Smoothing Forecast')
        plt.xlabel('Time Steps')
        plt.ylabel(data_dict['target_col'])
        plt.legend()
        plt.grid(True)
        plt.savefig('es_forecast.png')
        plt.show()
        
        # Visualize future forecast
        plt.figure(figsize=(12, 6))
        plt.plot(np.arange(len(train_series) + len(test_series)), 
                np.concatenate([train_series.values, test_series.values]), label='Historical Data')
        plt.plot(np.arange(len(train_series) + len(test_series), 
                          len(train_series) + len(test_series) + forecast_periods), 
                future_forecast, label='Future Forecast', color='red')
        plt.title(f'{forecast_periods}-Period ES Future Forecast')
        plt.xlabel('Time Steps')
        plt.ylabel(data_dict['target_col'])
        plt.legend()
        plt.grid(True)
        plt.savefig('es_future_forecast.png')
        plt.show()
        
        return {
            'model': model,
            'y_pred': y_pred,
            'metrics': {
                'rmse': rmse,
                'mae': mae,
                'r2': r2
            },
            'future_forecast': future_forecast
        }
    
    except Exception as e:
        print(f"Error in Exponential Smoothing: {e}")
        return None

def linear_regression_forecast(data_dict):
    """
    Perform Linear Regression forecasting
    
    Args:
        data_dict: Dictionary with prepared data
    
    Returns:
        Dictionary with model, predictions, and metrics
    """
    print("\nPerforming Linear Regression Forecasting:")
    
    X_train = data_dict['X_train_scaled']
    y_train = data_dict['y_train_scaled']
    X_test = data_dict['X_test_scaled']
    y_test = data_dict['y_test_scaled']
    
    try:
        model = LinearRegression()
        model.fit(X_train, y_train)
        
        # Make predictions
        y_pred = model.predict(X_test)
        
        # If data was scaled, inverse transform predictions
        if data_dict['target_scaler'] is not None:
            y_pred_original = data_dict['target_scaler'].inverse_transform(y_pred.reshape(-1, 1)).flatten()
            y_test_original = data_dict['y_test']
        else:
            y_pred_original = y_pred
            y_test_original = y_test
        
        # Calculate metrics
        rmse = np.sqrt(mean_squared_error(y_test_original, y_pred_original))
        mae = mean_absolute_error(y_test_original, y_pred_original)
        r2 = r2_score(y_test_original, y_pred_original)
        
        print(f"Linear Regression Metrics - RMSE: {rmse:.4f}, MAE: {mae:.4f}, RÂ²: {r2:.4f}")
        
        # Get feature importances (coefficients)
        feature_importances = pd.DataFrame({
            'Feature': data_dict['feature_cols'],
            'Coefficient': model.coef_
        }).sort_values('Coefficient', key=abs, ascending=False)
        
        print("\nTop 5 Feature Coefficients:")
        print(feature_importances.head(5))
        
        # Visualize results
        plt.figure(figsize=(12, 6))
        plt.scatter(y_test_original, y_pred_original, alpha=0.5)
        plt.plot([y_test_original.min(), y_test_original.max()], 
                [y_test_original.min(), y_test_original.max()], 'r--')
        plt.xlabel('Actual')
        plt.ylabel('Predicted')
        plt.title('Linear Regression: Actual vs Predicted')
        plt.grid(True)
        plt.savefig('linear_regression_prediction.png')
        plt.show()
        
        return {
            'model': model,
            'y_pred': y_pred_original,
            'metrics': {
                'rmse': rmse,
                'mae': mae,
                'r2': r2
            },
            'feature_importances': feature_importances
        }
    
    except Exception as e:
        print(f"Error in Linear Regression: {e}")
        return None

def train_gradient_boosting_models(data_dict):
    """
    Train and compare different gradient boosting models
    
    Args:
        data_dict: Dictionary with prepared data
    
    Returns:
        Dictionary with models, predictions, and metrics
    """
    print("\nTraining Gradient Boosting Models:")
    
    X_train = data_dict['X_train_scaled']
    y_train = data_dict['y_train_scaled']
    X_test = data_dict['X_test_scaled']
    y_test = data_dict['y_test_scaled']
    
    # Initialize models
    models = {
        'XGBoost': XGBRegressor(random_state=42),
        'LightGBM': LGBMRegressor(random_state=42),
        'CatBoost': CatBoostRegressor(random_state=42, verbose=0)
    }
    
    # Train and evaluate each model
    results = {}
    best_score = -np.inf
    best_model = None
    best_name = None
    
    for model_name, model in models.items():
        try:
            # Train model
            print(f"Training {model_name}...")
            model.fit(X_train, y_train)
            
            # Make predictions
            y_pred = model.predict(X_test)
            
            # If data was scaled, inverse transform predictions
            if data_dict['target_scaler'] is not None:
                y_pred_original = data_dict['target_scaler'].inverse_transform(y_pred.reshape(-1, 1)).flatten()
                y_test_original = data_dict['y_test']
            else:
                y_pred_original = y_pred
                y_test_original = y_test
            
            # Calculate metrics
            rmse = np.sqrt(mean_squared_error(y_test_original, y_pred_original))
            mae = mean_absolute_error(y_test_original, y_pred_original)
            r2 = r2_score(y_test_original, y_pred_original)
            evs = explained_variance_score(y_test_original, y_pred_original)
            
            # Cross-validation
            cv_scores = cross_val_score(
                model, X_train, y_train, 
                cv=min(5, len(X_train)//10 or 2),
                scoring='r2'
            )
            
            # Record results
            results[model_name] = {
                'model': model,
                'y_pred': y_pred_original,
                'RMSE': rmse, 
                'MAE': mae, 
                'RÂ²': r2,
                'Explained Variance': evs,
                'CV RÂ² Mean': np.mean(cv_scores),
                'CV RÂ² Std': np.std(cv_scores)
            }
            
            # Keep track of best model
            if r2 > best_score:
                best_score = r2
                best_model = model
                best_name = model_name
            
        except Exception as e:
            print(f"Error training {model_name}: {e}")
    
    # Create metrics DataFrame
    metrics_df = pd.DataFrame({
        model_name: {
            'RMSE': results[model_name]['RMSE'],
            'MAE': results[model_name]['MAE'],
            'RÂ²': results[model_name]['RÂ²'],
            'Explained Variance': results[model_name]['Explained Variance'],
            'CV RÂ² Mean': results[model_name]['CV RÂ² Mean'],
            'CV RÂ² Std': results[model_name]['CV RÂ² Std']
        } for model_name in results.keys()
    }).T
    
    metrics_df = metrics_df.sort_values('RÂ²', ascending=False)
    
    # Print results
    print("\nGradient Boosting Model Performance Metrics:")
    print(metrics_df)
    
    # Display feature importance for best model
    if best_model is not None:
        if hasattr(best_model, 'feature_importances_'):
            importances = best_model.feature_importances_
            feature_importance = pd.DataFrame({
                'Feature': data_dict['feature_cols'],
                'Importance': importances
            }).sort_values('Importance', ascending=False)
            
            print(f"\nTop 10 Feature Importances for {best_name}:")
            print(feature_importance.head(10))
            
            # Plot feature importance
            plt.figure(figsize=(12, 6))
            plt.barh(range(min(10, len(feature_importance))), 
                    feature_importance['Importance'].head(10), align='center')
            plt.yticks(range(min(10, len(feature_importance))), 
                      feature_importance['Feature'].head(10))
            plt.xlabel('Relative Importance')
            plt.title(f'Feature Importance ({best_name})')
            plt.tight_layout()
            plt.savefig('gb_feature_importance.png')
            plt.show()
            
            # Plot actual vs predicted
            plt.figure(figsize=(12, 6))
            plt.scatter(data_dict['y_test'], results[best_name]['y_pred'], alpha=0.5)
            plt.plot([data_dict['y_test'].min(), data_dict['y_test'].max()], 
                    [data_dict['y_test'].min(), data_dict['y_test'].max()], 'r--')
            plt.xlabel('Actual')
            plt.ylabel('Predicted')
            plt.title(f'{best_name}: Actual vs Predicted')
            plt.grid(True)
            plt.savefig(f'{best_name.lower()}_prediction.png')
            plt.show()
    
    return {
        'models': {name: results[name]['model'] for name in results},
        'predictions': {name: results[name]['y_pred'] for name in results},
        'metrics': metrics_df,
        'best_model': best_model,
        'best_model_name': best_name,
        'feature_importance': feature_importance if best_model is not None and hasattr(best_model, 'feature_importances_') else None
    }

def prepare_sequence_data(X, y, seq_length):
    """
    Prepare sequence data for LSTM/Transformer models
    
    Args:
        X: Input features
        y: Target values
        seq_length: Sequence length
    
    Returns:
        X_seq, y_seq
    """
    X_seq, y_seq = [], []
    for i in range(len(X) - seq_length):
        X_seq.append(X[i:i+seq_length])
        y_seq.append(y[i+seq_length])
    
    return np.array(X_seq), np.array(y_seq)

def build_lstm_model(input_shape):
    """Build an LSTM model for time series forecasting"""
    model = Sequential([
        LSTM(64, return_sequences=True, input_shape=input_shape),
        Dropout(0.2),
        LSTM(32),
        Dropout(0.2),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

def build_transformer_model(input_shape, head_size=32, num_heads=4, ff_dim=4):
    """Build a Transformer model for time series forecasting"""
    inputs = Input(shape=input_shape)
    x = inputs
    
    # Normalization and Attention
    x = LayerNormalization(epsilon=1e-6)(x)
    attention_output = MultiHeadAttention(
        num_heads=num_heads, key_dim=head_size
    )(x, x)
    x = tf.keras.layers.Add()([x, attention_output])
    x = LayerNormalization(epsilon=1e-6)(x)
    
    # Feed Forward Part
    ffn = Sequential([
        Dense(ff_dim * input_shape[-1], activation='relu'),
        Dense(input_shape[-1])
    ])
    ffn_output = ffn(x)
    x = tf.keras.layers.Add()([x, ffn_output])
    
    x = tf.keras.layers.GlobalAveragePooling1D()(x)
    outputs = Dense(1)(x)
    
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

def lstm_transformer_forecasting(data_dict, seq_length=12, forecast_periods=24):
    """
    Implement LSTM and Transformer forecasting
    
    Args:
        data_dict: Dictionary with prepared data
        seq_length: Sequence length for time series
        forecast_periods: Number of periods to forecast
    
    Returns:
        Dictionary with models, predictions, and metrics
    """
    print("\nPerforming LSTM and Transformer Forecasting:")
    
    X_train = data_dict['X_train_scaled']
    y_train = data_dict['y_train_scaled']
    X_test = data_dict['X_test_scaled']
    y_test = data_dict['y_test_scaled']
    
    # We need at least seq_length+1 samples
    if len(X_train) <= seq_length:
        print(f"Insufficient training data for sequence length {seq_length}")
        return None
    
    try:
        # Prepare sequence data
        X_train_seq, y_train_seq = prepare_sequence_data(X_train, y_train, seq_length)
        
        if len(X_test) > seq_length:
            X_test_seq, y_test_seq = prepare_sequence_data(X_test, y_test, seq_length)
        else:
            # If test set is smaller than sequence length, use the last seq_length elements from training + test
            X_combined = np.vstack([X_train[-seq_length+len(X_test):], X_test])
            y_combined = np.concatenate([y_train[-seq_length+len(X_test):], y_test])
            X_test_seq, y_test_seq = prepare_sequence_data(X_combined, y_combined, seq_length)
        
        # Verify we have data
        if len(X_train_seq) == 0 or len(X_test_seq) == 0:
            print("Insufficient data for sequence models after preparation")
            return None
        
        print(f"Sequence data shape - X_train: {X_train_seq.shape}, X_test: {X_test_seq.shape}")
        
        # Early stopping callback
        early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
        
        # Build and train LSTM model
        print("Training LSTM model...")
        lstm_model = build_lstm_model((seq_length, X_train.shape[1]))
        
        lstm_history = lstm_model.fit(
            X_train_seq, y_train_seq,
            epochs=50,
            batch_size=32,
            validation_split=0.2,
            callbacks=[early_stop],
            verbose=1
        )
        
        # Build and train Transformer model
        print("Training Transformer model...")
        transformer_model = build_transformer_model((seq_length, X_train.shape[1]))
        
        transformer_history = transformer_model.fit(
            X_train_seq, y_train_seq,
            epochs=50,
            batch_size=32,
            validation_split=0.2,
            callbacks=[early_stop],
            verbose=1
        )
        
        # Evaluate models
        lstm_pred = lstm_model.predict(X_test_seq)
        transformer_pred = transformer_model.predict(X_test_seq)
        
        # If data was scaled, inverse transform
        if data_dict['target_scaler'] is not None:
            lstm_pred_original = data_dict['target_scaler'].inverse_transform(lstm_pred).flatten()
            transformer_pred_original = data_dict['target_scaler'].inverse_transform(transformer_pred).flatten()
            y_test_seq_original = data_dict['target_scaler'].inverse_transform(y_test_seq.reshape(-1, 1)).flatten()
        else:
            lstm_pred_original = lstm_pred.flatten()
            transformer_pred_original = transformer_pred.flatten()
            y_test_seq_original = y_test_seq
        
        # Calculate metrics
        lstm_rmse = np.sqrt(mean_squared_error(y_test_seq_original, lstm_pred_original))
        lstm_mae = mean_absolute_error(y_test_seq_original, lstm_pred_original)
        lstm_r2 = r2_score(y_test_seq_original, lstm_pred_original)
        
        transformer_rmse = np.sqrt(mean_squared_error(y_test_seq_original, transformer_pred_original))
        transformer_mae = mean_absolute_error(y_test_seq_original, transformer_pred_original)
        transformer_r2 = r2_score(y_test_seq_original, transformer_pred_original)
        
        print(f"LSTM Metrics - RMSE: {lstm_rmse:.4f}, MAE: {lstm_mae:.4f}, RÂ²: {lstm_r2:.4f}")
        print(f"Transformer Metrics - RMSE: {transformer_rmse:.4f}, MAE: {transformer_mae:.4f}, RÂ²: {transformer_r2:.4f}")
        
        # Plot training history
        plt.figure(figsize=(12, 5))
        plt.subplot(1, 2, 1)
        plt.plot(lstm_history.history['loss'], label='Train Loss')
        plt.plot(lstm_history.history['val_loss'], label='Validation Loss')
        plt.title('LSTM Model Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        
        plt.subplot(1, 2, 2)
        plt.plot(transformer_history.history['loss'], label='Train Loss')
        plt.plot(transformer_history.history['val_loss'], label='Validation Loss')
        plt.title('Transformer Model Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.tight_layout()
        plt.savefig('dl_training_history.png')
        plt.show()
        
        # Plot predictions
        plt.figure(figsize=(12, 6))
        plt.plot(y_test_seq_original, label='Actual')
        plt.plot(lstm_pred_original, label='LSTM')
        plt.plot(transformer_pred_original, label='Transformer')
        plt.title('Deep Learning Models: Actual vs Predicted')
        plt.xlabel('Time Steps')
        plt.ylabel(data_dict['target_col'])
        plt.legend()
        plt.grid(True)
        plt.savefig('dl_predictions.png')
        plt.show()
        
        # Generate future forecasts with LSTM
        future_seq = X_train[-seq_length:] if len(X_train) >= seq_length else np.vstack([X_train, X_test])[-seq_length:]
        future_forecast_lstm = []
        future_forecast_transformer = []
        
        for _ in range(forecast_periods):
            # LSTM forecast
            next_pred_lstm = lstm_model.predict(future_seq.reshape(1, seq_length, X_train.shape[1]))
            future_forecast_lstm.append(next_pred_lstm[0, 0])
            
            # Transformer forecast
            next_pred_transformer = transformer_model.predict(future_seq.reshape(1, seq_length, X_train.shape[1]))
            future_forecast_transformer.append(next_pred_transformer[0, 0])
            
            # Create new input sequence by removing first element and adding prediction
            new_point = np.zeros((1, X_train.shape[1]))
            new_point[0, 0] = next_pred_lstm[0, 0]  # Use LSTM prediction as the input
            future_seq = np.vstack([future_seq[1:], new_point])
        
        # If data was scaled, inverse transform
        if data_dict['target_scaler'] is not None:
            future_forecast_lstm = data_dict['target_scaler'].inverse_transform(
                np.array(future_forecast_lstm).reshape(-1, 1)).flatten()
            future_forecast_transformer = data_dict['target_scaler'].inverse_transform(
                np.array(future_forecast_transformer).reshape(-1, 1)).flatten()
        
        # Plot future forecasts
        plt.figure(figsize=(12, 6))
        plt.plot(range(len(data_dict['y_train']) + len(data_dict['y_test'])), 
                np.concatenate([data_dict['y_train'], data_dict['y_test']]), 
                label='Historical Data')
        plt.plot(range(len(data_dict['y_train']) + len(data_dict['y_test']), 
                      len(data_dict['y_train']) + len(data_dict['y_test']) + forecast_periods), 
                future_forecast_lstm, label='LSTM Forecast', color='red')
        plt.plot(range(len(data_dict['y_train']) + len(data_dict['y_test']), 
                      len(data_dict['y_train']) + len(data_dict['y_test']) + forecast_periods), 
                future_forecast_transformer, label='Transformer Forecast', color='green')
        plt.title(f'{forecast_periods}-Period Future Forecast')
        plt.xlabel('Time Steps')
        plt.ylabel(data_dict['target_col'])
        plt.legend()
        plt.grid(True)
        plt.savefig('dl_future_forecast.png')
        plt.show()
        
        return {
            'models': {
                'lstm': lstm_model,
                'transformer': transformer_model
            },
            'predictions': {
                'lstm': lstm_pred_original,
                'transformer': transformer_pred_original
            },
            'metrics': {
                'lstm': {
                    'rmse': lstm_rmse,
                    'mae': lstm_mae,
                    'r2': lstm_r2
                },
                'transformer': {
                    'rmse': transformer_rmse,
                    'mae': transformer_mae,
                    'r2': transformer_r2
                }
            },
'future_forecast': {
                'lstm': future_forecast_lstm,
                'transformer': future_forecast_transformer
            }
        }
    
    except Exception as e:
        print(f"Error in LSTM/Transformer forecasting: {e}")
        return None

def compare_models(data_dict, model_results):
    """
    Compare all models and create ensemble predictions
    
    Args:
        data_dict: Dictionary with prepared data
        model_results: Dictionary with results from all models
    
    Returns:
        DataFrame with all metrics
    """
    print("\nComparing All Models:")
    
    models_to_compare = {}
    
    # Add Linear Regression results
    if 'linear_regression' in model_results and model_results['linear_regression'] is not None:
        models_to_compare['Linear Regression'] = {
            'RMSE': model_results['linear_regression']['metrics']['rmse'],
            'MAE': model_results['linear_regression']['metrics']['mae'],
            'RÂ²': model_results['linear_regression']['metrics']['r2']
        }
    
    # Add Exponential Smoothing results
    if 'exponential_smoothing' in model_results and model_results['exponential_smoothing'] is not None:
        models_to_compare['Exponential Smoothing'] = {
            'RMSE': model_results['exponential_smoothing']['metrics']['rmse'],
            'MAE': model_results['exponential_smoothing']['metrics']['mae'],
            'RÂ²': model_results['exponential_smoothing']['metrics']['r2']
        }
    
    # Add Gradient Boosting models results
    if 'gradient_boosting' in model_results and model_results['gradient_boosting'] is not None:
        for model_name in model_results['gradient_boosting']['metrics'].index:
            models_to_compare[model_name] = {
                'RMSE': model_results['gradient_boosting']['metrics'].loc[model_name, 'RMSE'],
                'MAE': model_results['gradient_boosting']['metrics'].loc[model_name, 'MAE'],
                'RÂ²': model_results['gradient_boosting']['metrics'].loc[model_name, 'RÂ²']
            }
    
    # Add Deep Learning models results
    if 'deep_learning' in model_results and model_results['deep_learning'] is not None:
        models_to_compare['LSTM'] = {
            'RMSE': model_results['deep_learning']['metrics']['lstm']['rmse'],
            'MAE': model_results['deep_learning']['metrics']['lstm']['mae'],
            'RÂ²': model_results['deep_learning']['metrics']['lstm']['r2']
        }
        models_to_compare['Transformer'] = {
            'RMSE': model_results['deep_learning']['metrics']['transformer']['rmse'],
            'MAE': model_results['deep_learning']['metrics']['transformer']['mae'],
            'RÂ²': model_results['deep_learning']['metrics']['transformer']['r2']
        }
    
    # Create a DataFrame with all metrics
    comparison_df = pd.DataFrame(models_to_compare).T
    comparison_df = comparison_df.sort_values('RÂ²', ascending=False)
    
    print("\nModel Performance Comparison:")
    print(comparison_df)
    
    # Visualize metrics comparison
    plt.figure(figsize=(14, 8))
    
    # RMSE
    plt.subplot(1, 3, 1)
    comparison_df['RMSE'].sort_values().plot(kind='barh')
    plt.title('RMSE (lower is better)')
    plt.grid(True)
    
    # MAE
    plt.subplot(1, 3, 2)
    comparison_df['MAE'].sort_values().plot(kind='barh')
    plt.title('MAE (lower is better)')
    plt.grid(True)
    
    # RÂ²
    plt.subplot(1, 3, 3)
    comparison_df['RÂ²'].sort_values(ascending=False).plot(kind='barh')
    plt.title('RÂ² (higher is better)')
    plt.grid(True)
    
    plt.tight_layout()
    plt.savefig('model_comparison.png')
    plt.show()
    
    # Create ensemble prediction
    predictions = []
    weights = []
    
    # For simplicity, use RÂ² as weight
    for model_name, metrics in models_to_compare.items():
        if 'Linear Regression' == model_name and 'linear_regression' in model_results:
            predictions.append(model_results['linear_regression']['y_pred'])
            weights.append(max(0.1, metrics['RÂ²']))
        elif 'Exponential Smoothing' == model_name and 'exponential_smoothing' in model_results:
            predictions.append(model_results['exponential_smoothing']['y_pred'])
            weights.append(max(0.1, metrics['RÂ²']))
        elif model_name in ['XGBoost', 'LightGBM', 'CatBoost'] and 'gradient_boosting' in model_results:
            predictions.append(model_results['gradient_boosting']['predictions'][model_name])
            weights.append(max(0.1, metrics['RÂ²']))
        elif model_name == 'LSTM' and 'deep_learning' in model_results:
            # Only include if we have matching samples
            if len(model_results['deep_learning']['predictions']['lstm']) == len(data_dict['y_test']):
                predictions.append(model_results['deep_learning']['predictions']['lstm'])
                weights.append(max(0.1, metrics['RÂ²']))
        elif model_name == 'Transformer' and 'deep_learning' in model_results:
            # Only include if we have matching samples
            if len(model_results['deep_learning']['predictions']['transformer']) == len(data_dict['y_test']):
                predictions.append(model_results['deep_learning']['predictions']['transformer'])
                weights.append(max(0.1, metrics['RÂ²']))
    
    # Only create ensemble if we have multiple predictions of same length
    if len(predictions) >= 2:
        # Check if all predictions have the same length
        if len(set([len(p) for p in predictions])) == 1:
            # Calculate weighted ensemble
            weights = np.array(weights) / np.sum(weights)  # Normalize weights
            ensemble_pred = np.zeros_like(predictions[0])
            
            for i, pred in enumerate(predictions):
                ensemble_pred += pred * weights[i]
            
            # Calculate ensemble metrics
            ensemble_rmse = np.sqrt(mean_squared_error(data_dict['y_test'], ensemble_pred))
            ensemble_mae = mean_absolute_error(data_dict['y_test'], ensemble_pred)
            ensemble_r2 = r2_score(data_dict['y_test'], ensemble_pred)
            
            print("\nEnsemble Model Performance:")
            print(f"RMSE: {ensemble_rmse:.4f}, MAE: {ensemble_mae:.4f}, RÂ²: {ensemble_r2:.4f}")
            
            # Add ensemble to comparison
            comparison_df.loc['Ensemble'] = {
                'RMSE': ensemble_rmse,
                'MAE': ensemble_mae,
                'RÂ²': ensemble_r2
            }
            
            # Visualize ensemble prediction
            plt.figure(figsize=(12, 6))
            plt.plot(data_dict['y_test'], label='Actual')
            plt.plot(ensemble_pred, label='Ensemble Prediction')
            plt.title('Ensemble Model: Actual vs Predicted')
            plt.xlabel('Time Steps')
            plt.ylabel(data_dict['target_col'])
            plt.legend()
            plt.grid(True)
            plt.savefig('ensemble_prediction.png')
            plt.show()
    
    return comparison_df

def generate_forecast_report(data_dict, model_results, comparison_df, forecast_periods=24):
    """
    Generate a summary report with forecasts from all models
    
    Args:
        data_dict: Dictionary with prepared data
        model_results: Dictionary with results from all models
        comparison_df: DataFrame with model comparisons
        forecast_periods: Number of periods to forecast
    """
    print("\nGenerating Forecast Report:")
    
    # Get the best model based on RÂ²
    best_model_name = comparison_df['RÂ²'].idxmax()
    print(f"Best performing model: {best_model_name} (RÂ² = {comparison_df.loc[best_model_name, 'RÂ²']:.4f})")
    
    # Get future forecasts from all available models
    future_forecasts = {}
    
    # Linear Regression - doesn't directly support future forecasting without new features
    
    # Exponential Smoothing
    if 'exponential_smoothing' in model_results and model_results['exponential_smoothing'] is not None:
        future_forecasts['Exponential Smoothing'] = model_results['exponential_smoothing']['future_forecast']
    
    # Deep Learning models
    if 'deep_learning' in model_results and model_results['deep_learning'] is not None:
        future_forecasts['LSTM'] = model_results['deep_learning']['future_forecast']['lstm']
        future_forecasts['Transformer'] = model_results['deep_learning']['future_forecast']['transformer']
    
    # Plot all future forecasts
    plt.figure(figsize=(12, 6))
    
    # Plot historical data
    plt.plot(range(len(data_dict['y_train']) + len(data_dict['y_test'])), 
            np.concatenate([data_dict['y_train'], data_dict['y_test']]), 
            label='Historical Data')
    
    # Plot forecasts
    for model_name, forecast in future_forecasts.items():
        plt.plot(range(len(data_dict['y_train']) + len(data_dict['y_test']), 
                      len(data_dict['y_train']) + len(data_dict['y_test']) + len(forecast)), 
                forecast, label=f'{model_name} Forecast')
    
    plt.title(f'{forecast_periods}-Period Future Forecast Comparison')
    plt.xlabel('Time Steps')
    plt.ylabel(data_dict['target_col'])
    plt.legend()
    plt.grid(True)
    plt.savefig('future_forecast_comparison.png')
    plt.show()
    
    # Create ensemble forecast if possible
    if len(future_forecasts) >= 2:
        ensemble_forecast = np.zeros(forecast_periods)
        weights = []
        
        for model_name, forecast in future_forecasts.items():
            if model_name in comparison_df.index:
                weights.append(max(0.1, comparison_df.loc[model_name, 'RÂ²']))
            else:
                weights.append(0.1)  # Default weight
        
        # Normalize weights
        weights = np.array(weights) / np.sum(weights)
        
        # Calculate weighted ensemble
        i = 0
        for model_name, forecast in future_forecasts.items():
            ensemble_forecast += forecast * weights[i]
            i += 1
        
        # Add ensemble forecast to the plot
        plt.figure(figsize=(12, 6))
        plt.plot(range(len(data_dict['y_train']) + len(data_dict['y_test'])), 
                np.concatenate([data_dict['y_train'], data_dict['y_test']]), 
                label='Historical Data')
        
        plt.plot(range(len(data_dict['y_train']) + len(data_dict['y_test']), 
                      len(data_dict['y_train']) + len(data_dict['y_test']) + len(ensemble_forecast)), 
                ensemble_forecast, label='Ensemble Forecast', linewidth=3)
        
        plt.title(f'{forecast_periods}-Period Ensemble Future Forecast')
        plt.xlabel('Time Steps')
        plt.ylabel(data_dict['target_col'])
        plt.legend()
        plt.grid(True)
        plt.savefig('ensemble_future_forecast.png')
        plt.show()
        
        print("\nEnsemble Future Forecast Values:")
        forecast_df = pd.DataFrame({
            'Period': range(1, forecast_periods + 1),
            'Forecast Value': ensemble_forecast
        })
        print(forecast_df)
    
    # Print a summary of the best models
    print("\nTop 3 Models Summary:")
    for model_name in comparison_df.sort_values('RÂ²', ascending=False).head(3).index:
        print(f"{model_name}: RMSE = {comparison_df.loc[model_name, 'RMSE']:.4f}, "
              f"MAE = {comparison_df.loc[model_name, 'MAE']:.4f}, "
              f"RÂ² = {comparison_df.loc[model_name, 'RÂ²']:.4f}")

def main(filepath, target_col='ptp_vel', forecast_periods=24):
    """
    Main function to run all forecasting models
    
    Args:
        filepath: Path to dataset
        target_col: Target column for forecasting
        forecast_periods: Number of periods to forecast
    """
    print("\n" + "="*50)
    print(f"TIME SERIES FORECASTING ANALYSIS: {target_col}")
    print("="*50)
    
    # Load and prepare data
    data_dict = load_and_prepare_data(filepath, target_col=target_col)
    
    if data_dict is None:
        print("Failed to load or prepare data. Exiting.")
        return
    
    # Dictionary to store model results
    model_results = {}
    
    # Run Exponential Smoothing
    model_results['exponential_smoothing'] = exponential_smoothing_forecast(
        data_dict, forecast_periods=forecast_periods
    )
    
    # Run Linear Regression
    model_results['linear_regression'] = linear_regression_forecast(data_dict)
    
    # Run Gradient Boosting models
    model_results['gradient_boosting'] = train_gradient_boosting_models(data_dict)
    
    # Run LSTM and Transformer
    model_results['deep_learning'] = lstm_transformer_forecasting(
        data_dict, forecast_periods=forecast_periods
    )
    
    # Compare all models
    comparison_df = compare_models(data_dict, model_results)
    
    # Generate forecast report
    generate_forecast_report(data_dict, model_results, comparison_df, forecast_periods=forecast_periods)
    
    print("\n" + "="*50)
    print("ANALYSIS COMPLETE")
    print("="*50)
    
    return {
        'data_dict': data_dict,
        'model_results': model_results,
        'comparison_df': comparison_df
    }

if __name__ == "__main__":
    # Define parameters
    filepath = "velocity_features.csv"  # Path to CSV file
    target_col = "ptp_vel"  # Target column to forecast
    forecast_periods = 24   # Number of periods to forecast
    
    # Run analysis
    results = main(filepath, target_col, forecast_periods)
```
